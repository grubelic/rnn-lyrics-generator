```python
def scraper():
    #artist = input('artist name  ')
    base = 'https://www.lyrics.com'
    artist = 'Franz Ferdinand'  
    url = base + '/artist/'
    
    if len(artist.split()) < 2 :
        url+= artist.split(' ')[0]
    else:
        name = artist.split(' ')
        url += name[0]
        for item in name[1:]:
            url = url + '-' + item
    print(url) 
    
    
#     for i in range(1, len(artist.split())):
#         url = url + '%20' + artist.split(' ')[i].lower()
#     print(url)
    r = requests.get(url)
    
    soup = BeautifulSoup(r.text, 'html.parser')
    alb = []
    s = []

    albums = soup.find_all('h3',{'class':'artist-album-label'})

    
    for i in albums:
        alb.append(i.text)
    print(alb)
    songs = soup.find_all('td',{'class':'tal qx'})
    for i in songs:
        s.append(i.text)
        
    l = []
    for i in range(len(songs)):
        if songs[i].a is None:
            pass
        else:
            url_lyr = base + songs[i].a.attrs['href']
            r = requests.get(url_lyr)
            soup = BeautifulSoup(r.text, 'html.parser')
            l.append(soup.find('pre',{'id':'lyric-body-text'}).text)
        
    df = pd.DataFrame({'song': s,'lyrics': l })
    return df
```
